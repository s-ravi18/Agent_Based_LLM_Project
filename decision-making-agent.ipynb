### Decision-Making Agent Prompt

````python
# Prompt for Decision Maker Agent
def classify_query(query):
    prompt = f"""
    You are a decision-making assistant for Company XYZ. Your role is to classify user queries into one of the following types:

    1. **General Query:** Can be answered using general knowledge and reasoning without referencing any company-specific data.
    2. **Technical Query:** Relates to project details, requiring access to the internal project database (e.g., techniques used, results, or project-specific knowledge).
    3. **Employee Details Query:** Pertains to employee-related information (e.g., name, email, role, salary, location, designation, or hiring details) and requires consulting the company's employee database.

    For each user query, return:
    - The query type (1, 2, or 3)
    - Reasoning for your classification

    **Examples:**

    **Example 1:**
    *User Query:* Which city has the highest average salary among employees?
    *Answer:*
    ```json
    {{
        "query_type": "3",
        "reasoning": "The query requires analyzing the employee database to group employees by location and calculate the average salary."
    }}
    ```

    **Example 2:**
    *User Query:* What is the latest AI technology trend?
    *Answer:*
    ```json
    {{
        "query_type": "1",
        "reasoning": "The query is general and can be answered using general knowledge without requiring access to company data."
    }}
    ```

    **Example 3:**
    *User Query:* What technologies were used in Project Alpha?
    *Answer:*
    ```json
    {{
        "query_type": "2",
        "reasoning": "This query pertains to a specific project and requires checking the internal project database for details about the technologies used."
    }}
    ```

     **Example 4:**
    *User Query:* How does economics work?
    *Answer:*
    ```json
    {{
        "query_type": "1",
        "reasoning": "The query is general and can be answered using general knowledge without requiring access to company data."
    }}
    ```

    **User Query:** {query}
    **Provide your classification in the below format only:**
    ```json
    {{
        "query_type": "1, 2, or 3",
        "reasoning": ""
    }}
    ```
    """
    return prompt


import json

def decode_classification_response(res):
    """
    Decodes the JSON response from the classification function.

    Parameters:
        json_response (str): JSON string containing query_type and reasoning.

    Returns:
        dict: A dictionary with the decoded query_type and reasoning.
    """
    try:
        if not isinstance(res, str):
            raise ValueError("Response is not a string.")
        
        # Clean up the response to isolate the JSON
        start_index = res.find("{")
        end_index = res.rfind("}") + 1
        if start_index == -1 or end_index == -1:
            raise ValueError("No valid JSON found in the response.")
        
        json_response = res[start_index:end_index]
        # print(f"Extracted JSON: {json_response}")
        
        response_dict = json.loads(json_response)
        
        # Validate keys
        if "query_type" not in response_dict or "reasoning" not in response_dict:
            raise ValueError("Invalid JSON structure: Missing 'query_type' or 'reasoning'.")
        
        # Extract values
        query_type = response_dict["query_type"]
        reasoning = response_dict["reasoning"]
        
        # Return the result as a dictionary
        return {
            "query_type": query_type,
            "reasoning": reasoning
        }
    
    except json.JSONDecodeError as e:
        raise ValueError(f"Invalid JSON format: {e}")
    except Exception as e:
        raise ValueError(f"An error occurred: {e}")
````

### SQL-Agent Prompt

````python
# SQl Agent

def sql_agent_query(user_query):
    prompt = f"""
    You are an expert SQL assistant. Your task is to generate SQL queries that work on an SQLite3 database based on the user-provided request. 

    The database contains the following table schema:

    **Employee Table:**
    - employee_name (TEXT, NOT NULL): Name of the employee.
    - employee_email (TEXT, UNIQUE, NOT NULL): Email address of the employee.
    - org_name (TEXT): Organization name the employee works for.
    - designation (TEXT, NOT NULL): Job designation of the employee.
    - years_of_experience (REAL, NOT NULL): Total years of professional experience.
    - salary (REAL): Current salary of the employee.
    - location (TEXT): Work location of the employee.
    - hire_date (TEXT): Date of hiring.

    ### Task:
    Generate an **SQLite3-compatible SQL query** that fulfills the user's request. 
    - Do not include explanatory comments in the generated SQL.
    - Ensure the SQL is syntactically correct and adheres to the provided schema.
    - Do not modify the schema or add fields.

    ### Examples:
    **Example 1:**
    *User Query:* Get the average salary of employees grouped by their job designation.
    *Output:*
    ```json
    {{
        "query": "SELECT designation, AVG(salary) AS average_salary FROM Employee GROUP BY designation;"
    }}
    ```

    **Example 2:**
    *User Query:* Retrieve the names and email addresses of employees hired in the last 5 years.
    *Output:*
    ```json
    {{
        "query": "SELECT employee_name, employee_email FROM Employee WHERE hire_date >= DATE('now', '-5 years');"
    }}
    ```

    ### User Query:
    {user_query}

    ### Output Format:
    Provide the result in the following JSON format:
    ```json
    {{
        "query": "<Generated SQL query>"
    }}
    ```

    """
    return prompt

def sql_query_result(res, cursor):
    """
    Decodes the JSON response to extract a SQL query and executes it.

    Parameters:
        res (str): The raw response from the LLM API.
        cursor: Database cursor object for executing the query.

    Returns:
        list: A list of rows resulting from the executed query.
    """
    try:
        # Debugging: Check the response format
        # print(f"Type of res: {type(res)}")
        # print(f"Content of res: {res}")

        # Extract JSON portion from the response
        if not isinstance(res, str):
            raise ValueError("Response is not a string.")
        
        # Clean up the response to isolate the JSON
        start_index = res.find("{")
        end_index = res.rfind("}") + 1
        if start_index == -1 or end_index == -1:
            raise ValueError("No valid JSON found in the response.")
        
        json_response = res[start_index:end_index]
        # print(f"Extracted JSON: {json_response}")

        # Parse the JSON response
        response_dict = json.loads(json_response)
        
        # Validate that the query exists
        if "query" not in response_dict:
            raise ValueError("Invalid JSON structure: Missing 'query'.")
        
        # Extract and execute the SQL query
        sql_query = response_dict["query"].strip()
        if not sql_query:
            raise ValueError("Empty SQL query in the response.")

        print(f"Executing SQL Query: {sql_query}")
        cursor.execute(sql_query)
        rows = cursor.fetchall()

        # Return the query result
        return rows

    except json.JSONDecodeError as e:
        raise ValueError(f"Invalid JSON format: {e}")
    except Exception as e:
        raise ValueError(f"An error occurred: {e}")
````

## RAG-Agent Query

````python
from langchain.prompts import PromptTemplate

def RAG_agent(query, custom_llm):
    # Define a custom prompt for the question-answering system
    prompt_template = """
    You are a helpful assistant for Company XYZ. You have access to the following project documents:
    
    {context}
    
    Based on the information from these documents, please answer the following question:
    
    {question}
    """
    
    # Create a prompt template instance
    prompt = PromptTemplate(
        input_variables=["context", "question"],
        template=prompt_template
    )
    
    # Use the prompt in the RetrievalQA chain
    qa_chain = RetrievalQA.from_chain_type(
        llm=custom_llm,
        retriever=knowledge_base.as_retriever(search_kwargs={"k": 5}),
        chain_type_kwargs={"prompt": prompt}
    )
    
    # Ask a question
    # query = "'Have we done anything on Air Quality improvement?',"
    response = qa_chain.run(query)

    ## Alternative syntax;
    ## response = qa_chain.run({"question": query}) 

    return response
````
